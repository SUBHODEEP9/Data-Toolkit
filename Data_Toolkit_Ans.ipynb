1. What is NumPy, and why is it widely used in Python?
  * NumPy is a Python library used for numerical and scientific computing because it provides fast and efficient array operations.
2. How does broadcasting work in NumPy?
  * Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes without manually resizing them.

```python
import numpy as np

arr = np.array([1, 2, 3])
print(arr + 5)
```

3. What is a Pandas DataFrame?
  * A Pandas DataFrame is a two-dimensional table-like data structure used for storing and analyzing data in rows and columns.

```python
import pandas as pd

data = pd.DataFrame({
    "Name": ["Amit", "Riya"],
    "Age": [21, 22]
})
print(data)
```

4. Explain the use of the groupby() method in Pandas.
  * The groupby() method is used to group data based on one or more columns and perform aggregate operations like sum or average.

```python
import pandas as pd

df = pd.DataFrame({
    "Dept": ["IT", "IT", "HR"],
    "Salary": [5000, 6000, 4000]
})

print(df.groupby("Dept").sum())
```

5. Why is Seaborn preferred for statistical visualizations?
  * Seaborn is preferred because it provides attractive statistical graphs with less code and better default styles.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.histplot(tips["total_bill"])
plt.show()
```

6. What are the differences between NumPy arrays and Python lists?
  * NumPy arrays are faster, consume less memory, and support mathematical operations directly, while Python lists are more flexible but slower.

```python
import numpy as np

arr = np.array([1, 2, 3])
print(arr + 2)
```

7. What is a heatmap, and when should it be used?
A heatmap is a graphical representation of data where values are shown using colors, and it is commonly used to display correlations.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [[1, 2], [3, 4]]
sns.heatmap(data)
plt.show()
```

8. What does the term “vectorized operation” mean in NumPy?
  * A vectorized operation means performing operations on entire arrays without using loops, making computations faster.

```python
import numpy as np

arr = np.array([1, 2, 3])
print(arr * 10)
```

9. How does Matplotlib differ from Plotly?
  * Matplotlib creates static graphs, while Plotly creates interactive and web-friendly visualizations.

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3], [4,5,6])
plt.show()
```

10. What is the significance of hierarchical indexing in Pandas?
  * Hierarchical indexing allows multiple levels of indexing in a DataFrame, making complex data analysis easier.

```python
import pandas as pd

data = pd.Series([100, 200], index=[["A", "A"], ["Math", "Science"]])
print(data)
```

11. What is the role of Seaborn’s pairplot() function?
  * The pairplot() function is used to visualize relationships between multiple variables in a dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.show()
```

12. What is the purpose of the describe() function in Pandas?
  * The describe() function generates summary statistics such as mean, count, minimum, and maximum values of data.

```python
import pandas as pd

df = pd.DataFrame({"Marks": [70, 80, 90]})
print(df.describe())
```

13. Why is handling missing data important in Pandas?
  * Handling missing data is important because incomplete data can produce incorrect analysis and results.

```python
import pandas as pd

df = pd.DataFrame({"Marks": [90, None, 80]})
print(df.fillna(0))
```

15. How does NumPy handle multidimensional arrays?

17. Explain the difference between apply() and map() in Pandas.
  * map() is used mainly for element-wise operations on a Series.
  * while apply() can operate on rows, columns, or entire DataFrames.

```python
import pandas as pd

s = pd.Series([1,2,3])
print(s.map(lambda x: x*2))
```

18. What are some advanced features of NumPy?
  * Advanced NumPy features include broadcasting, multidimensional arrays, linear algebra functions, and random number generation.

```python
import numpy as np

a = np.array([[1,2], [3,4]])
print(np.linalg.det(a))
```

19. How does Pandas simplify time series analysis?
  * Pandas simplifies time series analysis by providing date handling, resampling, and rolling window operations.

```python
import pandas as pd

dates = pd.date_range("2025-01-01", periods=3)
print(dates)
```

20. What is the role of a pivot table in Pandas?
  * A pivot table summarizes and reorganizes data for easier analysis.

```python
import pandas as pd

df = pd.DataFrame({
    "Dept": ["IT", "HR"],
    "Salary": [5000, 4000]
})

print(pd.pivot_table(df, values="Salary", index="Dept"))
```

21. Why is NumPy’s array slicing faster than Python’s list slicing?
  * NumPy array slicing is faster because arrays store data in contiguous memory locations and use optimized C code internally.

```python
import numpy as np

arr = np.array([1,2,3,4])
print(arr[1:3])
```

```python
# <A How do you create a 2D NumPy array and calculate the sum of each row)

import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = arr.sum(axis=1)

print("Array:")
print(arr)

print("Sum of each row:")
print(row_sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd

df = pd.DataFrame({
    "Marks": [70, 85, 90, 60]
})

mean_value = df["Marks"].mean()

print("Mean Marks:", mean_value)
```

```python
# 4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "A": [1,2,3],
    "B": [4,5,6]
})

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np

arr1 = np.array([2,5,6])
arr2 = np.array([3,6,9])

result = arr1 * arr2
print("Result of 2 Numpy array multiplication :", result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]

y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

plt.plot(x, y1, label="Square")
plt.plot(x, y2, label="Linear")

plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend()

plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd

df = pd.DataFrame({
    "Marks": [55, 60, 65, 72, 70, 85, 95]
})

threshold = 68
filtered_df = df[df["Marks"] > threshold]

print("Filtered DataFrame:")
print(filtered_df)
```

```python
# 11. Perform matrix multiplication using NumPy.

import numpy as np

a = np.array([[1,2],
              [7, 8]])
b = np.array([[3, 4],
              [5,6]])

result = a @ b

print("Matrix multiplication result :", result)
```
